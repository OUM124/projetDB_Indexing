```python
import heapq
import os
import numpy as np
from graphviz import Digraph
import math

class KDNode:
    def __init__(self, points, axis):
        self.points = points  # Array of points for leaf nodes
        self.axis = axis     # Splitting axis
        self.left = None     # Left child
        self.right = None    # Right child
        self.split_value = None  # Split value for internal nodes

class KDTree:
    def __init__(self, points=None, leafsize=16, discriminator="cyclic"):
        self.root = None
        self.leafsize = leafsize
        self.discriminator = discriminator
        if points is not None:
            points = np.asarray(points)
            self.root = self.kdtree(points)

    def kdtree(self, point_array, depth=0):
        if point_array.shape[0] == 0:
            return None

        # Create leaf node if points <= leafsize
        if point_array.shape[0] <= self.leafsize:
            axis = depth % point_array.shape[1]
            return KDNode(points=point_array, axis=axis)

        # Choose axis based on strategy
        if self.discriminator == "cyclic":
            axis = depth % point_array.shape[1]
        elif self.discriminator == "variance":
            axis = np.argmax(np.var(point_array, axis=0))
        else:
            raise ValueError("Invalid discriminator strategy")

        # Sort and split points
        idx = np.argsort(point_array[:, axis])
        sorted_points = point_array[idx]
        median_idx = sorted_points.shape[0] // 2

        # Create internal node
        node = KDNode(points=None, axis=axis)
        node.split_value = sorted_points[median_idx, axis]
        node.left = self.kdtree(sorted_points[:median_idx], depth + 1)
        node.right = self.kdtree(sorted_points[median_idx:], depth + 1)
        return node

    def insert(self, point):
        point = np.asarray(point)

        def _insert(node, point, depth):
            if node is None:
                return KDNode(points=point.reshape(1, -1), axis=depth % point.shape[0])

            if node.points is not None:
                # If leaf node still has space
                if len(node.points) < self.leafsize:
                    node.points = np.vstack([node.points, point])
                    return node

                # Split leaf node if full
                all_points = np.vstack([node.points, point])
                node = self.kdtree(all_points, depth)
                return node

            # Internal node
            if point[node.axis] < node.split_value:
                node.left = _insert(node.left, point, depth + 1)
            else:
                node.right = _insert(node.right, point, depth + 1)
            return node

        self.root = _insert(self.root, point, 0)

    def nearest_neighbor(self, query_point):
        query_point = np.asarray(query_point)

        def squared_distance(p1, p2):
            return np.sum((p1 - p2) ** 2)

        def _search(node):
            if node is None:
                return None, float('inf')

            if node.points is not None:
                # Leaf node: find nearest among points
                distances = np.sum((node.points - query_point) ** 2, axis=1)
                nearest_idx = np.argmin(distances)
                return node.points[nearest_idx], distances[nearest_idx]

            # Internal node
            if query_point[node.axis] < node.split_value:
                near, far = node.left, node.right
            else:
                near, far = node.right, node.left

            # Search near subtree
            best_point, best_dist = _search(near)

            # Check if we need to search far subtree
            if (query_point[node.axis] - node.split_value) ** 2 < best_dist:
                far_point, far_dist = _search(far)
                if far_dist < best_dist:
                    best_point, best_dist = far_point, far_dist

            return best_point, best_dist

        point, dist = _search(self.root)
        return point, math.sqrt(dist)

    def k_nearest_neighbors(self, query_point, k):
        query_point = np.asarray(query_point)

        def _search(node, heap):
            if node is None:
                return

            if node.points is not None:
                # Process all points in leaf
                dists = np.sum((node.points - query_point) ** 2, axis=1)
                for dist, point in zip(dists, node.points):
                    if len(heap) < k:
                        heapq.heappush(heap, (-dist, tuple(point)))
                    elif dist < -heap[0][0]:
                        heapq.heappushpop(heap, (-dist, tuple(point)))
                return

            # Internal node
            if query_point[node.axis] < node.split_value:
                near, far = node.left, node.right
            else:
                near, far = node.right, node.left

            _search(near, heap)

            if len(heap) < k or (query_point[node.axis] - node.split_value) ** 2 < -heap[0][0]:
                _search(far, heap)

        heap = []
        _search(self.root, heap)
        return [np.array(point) for _, point in sorted(heap, reverse=True)]

    def range_search(self, query_range):
        def _in_range(points, query_range):
            return np.all((points >= [lower for lower, _ in query_range]) &
                         (points <= [upper for _, upper in query_range]), axis=1)

        def _search(node):
            if node is None:
                return []

            points_in_range = []

            if node.points is not None:
                # Leaf node: check all points
                mask = _in_range(node.points, query_range)
                points_in_range.extend(node.points[mask])
                return points_in_range

            # Internal node
            axis = node.axis
            lower, upper = query_range[axis]

            if lower <= node.split_value:
                points_in_range.extend(_search(node.left))
            if upper >= node.split_value:
                points_in_range.extend(_search(node.right))

            return points_in_range

        return _search(self.root)

    def draw_kdtree(self, node=None, graph=None, parent_id=None, depth=0, output_filename="kdtree"):
        if graph is None:
            graph = Digraph(format='png')
            graph.attr('node', shape='circle')
            os.makedirs("images", exist_ok=True)

        if node is None:
            node = self.root or graph

        node_id = str(id(node))

        if node.points is not None:
            points_str = "\n".join([str(p) for p in node.points[:3]])
            if len(node.points) > 3:
                points_str += f"\n... ({len(node.points)} total)"
            label = f"Leaf Node\n{points_str}\nAxis: {node.axis}"
        else:
            label = f"Split: {node.split_value:.2f}\nAxis: {node.axis}"

        graph.node(node_id, label)

        if parent_id:
            graph.edge(parent_id, node_id)

        if node.left:
            self.draw_kdtree(node.left, graph, node_id, depth + 1)
        if node.right:
            self.draw_kdtree(node.right, graph, node_id, depth + 1)

        if parent_id is None:
            output_path = os.path.join("images", output_filename)
            graph.render(output_path, cleanup=True)
            print(f"KDTree image saved at: {output_path}.png")

        return graph




```

```python
import numpy as np
from collections import defaultdict

def create_vector_to_index_mapping(train_data, decimals=10):
    """
    Create a mapping from vector tuple to index for fast lookup.

    Args:
        train_data: numpy array of shape (1000000, 960)
        decimals: number of decimal places to round to
    """
    # Round the data to handle floating point precision
    rounded_data = np.round(train_data, decimals=decimals)

    # Create mapping of vector to index
    vector_to_idx = {}
    for idx, vector in enumerate(rounded_data):
        vector_to_idx[tuple(vector)] = idx

    return vector_to_idx

def batch_find_indices(vectors, vector_to_idx_map, decimals=10):
    """
    Find indices for multiple vectors at once using the precomputed mapping.

    Args:
        vectors: numpy array of vectors
        vector_to_idx_map: dict mapping vector tuples to indices
        decimals: number of decimal places to round to
    """
    rounded_vectors = np.round(vectors, decimals=decimals)
    indices = []
    for v in rounded_vectors:
        indices.append(vector_to_idx_map.get(tuple(v), -1))
    return np.array(indices)

def evaluate_knn_accuracy_optimized(knn_results, ground_truth_indices, train_data):
    """
    Optimized version of KNN accuracy evaluation.

    Args:
        knn_results: numpy array of shape (num_queries, num_neighbors) containing neighbor vectors
        ground_truth_indices: numpy array of shape (num_queries, num_neighbors) containing true indices
        train_data: numpy array of shape (num_train, vector_dim) containing training data
    """
    # Create vector to index mapping once
    vector_to_idx_map = create_vector_to_index_mapping(train_data)

    num_queries = len(knn_results)
    num_neighbors = len(knn_results[0])

    # Process each query point
    predicted_indices = []
    for query_neighbors in knn_results:
        # Find indices for all neighbors of this query point
        query_indices = batch_find_indices(query_neighbors, vector_to_idx_map)
        predicted_indices.append(query_indices)

    predicted_indices = np.array(predicted_indices)

    # Compute accuracy using vectorized operations
    correct_predictions = np.sum([
        np.isin(ground_truth_indices[i], predicted_indices[i]).sum()
        for i in range(num_queries)
    ])

    accuracy = correct_predictions / (num_queries * num_neighbors)
    return accuracy
```

```python
import numpy as np
import time
import psutil
import os
import matplotlib.pyplot as plt
from memory_profiler import profile
import gc

def measure_memory_usage():
    """Measure current memory usage of the process"""
    process = psutil.Process(os.getpid())
    return process.memory_info().rss / 1024 / 1024  # Convert to MB
def create_range_queries(query_points, range_size=0.1):
    """Create range queries from test points"""
    queries = []
    for point in query_points:
        # Create range query with fixed size around test point
        query_range = [(max(0, p - range_size/2), min(1, p + range_size/2))
                      for p in point]
        queries.append(query_range)
    return queries

def benchmark_kdtree(train_data, query_data, leaf_sizes, k=100, num_queries=None):
    """
    Benchmark KDTree performance with different leaf sizes

    Args:
        train_data: Training dataset
        query_data: Query dataset
        leaf_sizes: List of leaf sizes to test
        k: Number of nearest neighbors to find
        num_queries: Number of query points to test (None for all points)
    """
    results = {
        'construction_time': [],
        'memory_usage': [],
        'knn_query_time': [],
        'avg_knn_query_time': [],
        'range_query_time': [],
        'avg_range_query_time': []
    }

    # Select subset of query points if specified
    query_points = query_data.values
    if num_queries is not None:
        np.random.seed(42)  # For reproducibility
        indices = np.random.choice(len(query_points), num_queries, replace=False)
        query_points = query_points[indices]

    # Create range queries from the same test points
    range_queries = create_range_queries(query_points)

    print(f"\nStarting benchmark with {len(query_points)} query points...")

    for leaf_size in leaf_sizes:
        print(f"\nTesting leaf size: {leaf_size}")

        # Clear memory and garbage collect
        gc.collect()
        base_memory = measure_memory_usage()

        # Measure construction time and memory
        start_time = time.time()
        tree = KDTree(points=train_data, leafsize=leaf_size, discriminator="cyclic")
        construction_time = time.time() - start_time

        # Measure memory after tree construction
        tree_memory = measure_memory_usage() - base_memory

        # Measure kNN query time
        knn_query_times = []
        knn_start_time = time.time()

        for i, query_point in enumerate(query_points):
            query_start = time.time()
            _ = tree.k_nearest_neighbors(query_point, k)
            knn_query_times.append(time.time() - query_start)

            if (i + 1) % 100 == 0:
                print(f"Processed {i + 1}/{len(query_points)} kNN queries")

        total_knn_time = time.time() - knn_start_time

        # Measure range query time
        range_query_times = []
        range_start_time = time.time()

        for i, query_range in enumerate(range_queries):
            query_start = time.time()
            _ = tree.range_search(query_range)
            range_query_times.append(time.time() - query_start)

            if (i + 1) % 100 == 0:
                print(f"Processed {i + 1}/{len(range_queries)} range queries")

        total_range_time = time.time() - range_start_time

        # Store results
        results['construction_time'].append(construction_time)
        results['memory_usage'].append(tree_memory)
        results['knn_query_time'].append(total_knn_time)
        results['avg_knn_query_time'].append(np.mean(knn_query_times))
        results['range_query_time'].append(total_range_time)
        results['avg_range_query_time'].append(np.mean(range_query_times))

        print(f"\nLeaf size {leaf_size} results:")
        print(f"Construction time: {construction_time:.2f} seconds")
        print(f"Memory usage: {tree_memory:.2f} MB")
        print(f"Total kNN query time: {total_knn_time:.2f} seconds")
        print(f"Average kNN query time: {np.mean(knn_query_times):.6f} seconds")
        print(f"Total range query time: {total_range_time:.2f} seconds")
        print(f"Average range query time: {np.mean(range_query_times):.6f} seconds")

    return results
```

```python
def plot_results(leaf_sizes, results):
    """Plot benchmark results"""
    plt.figure(figsize=(15, 10))

    # Construction time plot
    plt.subplot(2, 3, 1)
    plt.plot(leaf_sizes, results['construction_time'], marker='o')
    plt.xlabel('Leaf Size')
    plt.ylabel('Construction Time (s)')
    plt.title('Construction Time vs Leaf Size')
    plt.grid(True)

    # Memory usage plot
    plt.subplot(2, 3, 2)
    plt.plot(leaf_sizes, results['memory_usage'], marker='o', color='green')
    plt.xlabel('Leaf Size')
    plt.ylabel('Memory Usage (MB)')
    plt.title('Memory Usage vs Leaf Size')
    plt.grid(True)

    # kNN query time plot
    plt.subplot(2, 3, 3)
    plt.plot(leaf_sizes, results['avg_knn_query_time'], marker='o', color='red')
    plt.xlabel('Leaf Size')
    plt.ylabel('Avg kNN Query Time (s)')
    plt.title('Average kNN Query Time vs Leaf Size')
    plt.grid(True)

    # Range query time plot
    plt.subplot(2, 3, 4)
    plt.plot(leaf_sizes, results['avg_range_query_time'], marker='o', color='purple')
    plt.xlabel('Leaf Size')
    plt.ylabel('Avg Range Query Time (s)')
    plt.title('Average Range Query Time vs Leaf Size')
    plt.grid(True)

    # Total query time comparison
    plt.subplot(2, 3, 5)
    plt.plot(leaf_sizes, results['knn_query_time'], marker='o', label='kNN')
    plt.plot(leaf_sizes, results['range_query_time'], marker='o', label='Range')
    plt.xlabel('Leaf Size')
    plt.ylabel('Total Query Time (s)')
    plt.title('Total Query Time Comparison')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.savefig('kdtree_performance_comparison.png')
    plt.close()
```

```python
def run_leaf_size_benchmark(train_data, query_data, num_queries=1000):
    """Main function to run the complete benchmark"""
    leaf_sizes = [16, 32, 64,  256]

    print("Starting KD-tree leaf size benchmark...")
    print(f"Training data shape: {train_data.shape}")
    print(f"Query data shape: {query_data.shape}")
    print(f"Testing leaf sizes: {leaf_sizes}")

    results = benchmark_kdtree(
        train_data=train_data,
        query_data=query_data,
        leaf_sizes=leaf_sizes,
        k=100,
        num_queries=num_queries
    )

    # Plot results
    plot_results(leaf_sizes, results)

    # Print summary
    print("\nBenchmark Summary:")
    print("\nLeaf Size | Const. Time (s) | Memory (MB) | Avg kNN Time (s) | Avg Range Time (s)")
    print("-" * 80)
    for i, leaf_size in enumerate(leaf_sizes):
        print(f"{leaf_size:^9} | {results['construction_time'][i]:^14.2f} | "
              f"{results['memory_usage'][i]:^11.2f} | {results['avg_knn_query_time'][i]:^15.6f} | "
              f"{results['avg_range_query_time'][i]:^16.6f}")

    return results
```

Benchmarking OF the discriminator variants

```python
import numpy as np
import time
import psutil
import os
import matplotlib.pyplot as plt
from memory_profiler import profile
import gc

def measure_memory_usage():
    """Measure current memory usage of the process"""
    process = psutil.Process(os.getpid())
    return process.memory_info().rss / 1024 / 1024  # Convert to MB

def benchmark_discriminators(train_data, query_data, k=100, num_queries=None):
    """
    Benchmark KDTree performance with different discriminator strategies

    Args:
        train_data: Training dataset
        query_data: Query dataset
        k: Number of nearest neighbors to find
        num_queries: Number of query points to test (None for all points)
    """
    results = {
        'discriminator': [],
        'construction_time': [],
        'memory_usage': [],
        'total_query_time': [],
        'avg_query_time': [],
        'query_times': []  # Store all individual query times for distribution analysis
    }

    # Select subset of query points if specified
    query_points = query_data.values
    if num_queries is not None:
        np.random.seed(42)  # For reproducibility
        indices = np.random.choice(len(query_points), num_queries, replace=False)
        query_points = query_points[indices]

    discriminators = ["cyclic", "variance"]

    print(f"\nStarting benchmark with {len(query_points)} query points...")

    for discriminator in discriminators:
        print(f"\nTesting discriminator: {discriminator}")

        # Clear memory and garbage collect
        gc.collect()
        base_memory = measure_memory_usage()

        # Measure construction time and memory
        start_time = time.time()
        tree = KDTree(points=train_data, discriminator=discriminator)
        construction_time = time.time() - start_time

        # Measure memory after tree construction
        tree_memory = measure_memory_usage() - base_memory

        # Measure query time
        query_times = []
        query_start_time = time.time()

        for i, query_point in enumerate(query_points):
            query_start = time.time()
            _ = tree.k_nearest_neighbors(query_point, k)
            query_times.append(time.time() - query_start)

            if (i + 1) % 100 == 0:
                print(f"Processed {i + 1}/{len(query_points)} queries")

        total_query_time = time.time() - query_start_time

        # Store results
        results['discriminator'].append(discriminator)
        results['construction_time'].append(construction_time)
        results['memory_usage'].append(tree_memory)
        results['total_query_time'].append(total_query_time)
        results['avg_query_time'].append(np.mean(query_times))
        results['query_times'].append(query_times)

        print(f"\n{discriminator.capitalize()} discriminator results:")
        print(f"Construction time: {construction_time:.2f} seconds")
        print(f"Memory usage: {tree_memory:.2f} MB")
        print(f"Total query time: {total_query_time:.2f} seconds")
        print(f"Average query time: {np.mean(query_times):.6f} seconds")
        print(f"Query time std dev: {np.std(query_times):.6f} seconds")

    return results
```

```python

def plot_comparison(results):
    """Plot benchmark comparison results"""
    plt.figure(figsize=(15, 10))

    # Bar width for grouped bars
    width = 0.35
    discriminators = results['discriminator']
    x = np.arange(len(discriminators))

    # Construction time comparison
    plt.subplot(2, 2, 1)
    plt.bar(x, results['construction_time'], width, label='Construction Time')
    plt.xlabel('Discriminator')
    plt.ylabel('Time (s)')
    plt.title('Construction Time Comparison')
    plt.xticks(x, discriminators)
    plt.grid(True)

    # Memory usage comparison
    plt.subplot(2, 2, 2)
    plt.bar(x, results['memory_usage'], width, color='green')
    plt.xlabel('Discriminator')
    plt.ylabel('Memory (MB)')
    plt.title('Memory Usage Comparison')
    plt.xticks(x, discriminators)
    plt.grid(True)

    # Average query time comparison
    plt.subplot(2, 2, 3)
    plt.bar(x, results['avg_query_time'], width, color='red')
    plt.xlabel('Discriminator')
    plt.ylabel('Time (s)')
    plt.title('Average Query Time Comparison')
    plt.xticks(x, discriminators)
    plt.grid(True)

    # Query time distribution
    plt.subplot(2, 2, 4)
    plt.boxplot([times for times in results['query_times']], labels=discriminators)
    plt.ylabel('Time (s)')
    plt.title('Query Time Distribution')
    plt.grid(True)

    plt.tight_layout()
    plt.savefig('kdtree_discriminator_comparison.png')
    plt.close()
```

```python

def run_discriminator_benchmark(train_data, query_data, num_queries=1000):
    """Main function to run the complete benchmark"""
    print("Starting KD-tree discriminator benchmark...")
    print(f"Training data shape: {train_data.shape}")
    print(f"Query data shape: {query_data.shape}")

    results = benchmark_discriminators(
        train_data=train_data,
        query_data=query_data,
        k=100,
        num_queries=num_queries
    )

    # Plot results
    plot_comparison(results)

    # Print summary
    print("\nBenchmark Summary:")
    print("\nDiscriminator | Const. Time (s) | Memory (MB) | Avg Query Time (s)")
    print("-" * 65)
    for i, disc in enumerate(results['discriminator']):
        print(f"{disc:^12} | {results['construction_time'][i]:^14.2f} | "
              f"{results['memory_usage'][i]:^11.2f} | {results['avg_query_time'][i]:^16.6f}")

    return results
```

Benchmarking of Performance between Saved/loaded VS rebuilded tree.

```python
import numpy as np
import time
import pickle
import matplotlib.pyplot as plt
import gc
import os

def save_tree_to_disk(tree, filename):
    """Save a KDTree to disk."""
    try:
        with open(filename, "wb") as file:
            pickle.dump(tree, file)
    except Exception as e:
        print(f"Error saving tree: {e}")
        return False
    return True

def load_tree_from_disk(filename):
    """Load a KDTree from disk."""
    try:
        with open(filename, "rb") as file:
            return pickle.load(file)
    except Exception as e:
        print(f"Error loading tree: {e}")
        return None

def benchmark_single_tree(train_data, query_data, leaf_size=16, k_values=None, num_queries=1000, discriminator="cyclic"):
    """
    Benchmark single KDTree configuration: loading vs rebuilding.
    Compare kNN query performance for different k values.

    Args:
        train_data: numpy array of training data
        query_data: numpy array of query data
        leaf_size: leaf size for KDTree construction
        k_values: list of k values to test for nearest neighbors
        num_queries: number of query points to use
        discriminator: strategy for choosing split axis ("cyclic" or "variance")

    Returns:
        dict: Dictionary containing benchmark results
    """
    if k_values is None:
        k_values = [1, 5, 10, 50]

    results = {
        'k_values': k_values,
        'rebuild_times': [],      # Total time (rebuild + query)
        'load_times': [],         # Total time (load + query)
        'rebuild_query_times': [], # Query times only
        'load_query_times': [],    # Query times only
        'memory_usage': []         # Memory usage for each k
    }

    # Ensure data is properly shaped
    train_data = np.asarray(train_data)
    query_data = np.asarray(query_data)

    if train_data.ndim == 1:
        train_data = train_data.reshape(-1, 1)
    if query_data.ndim == 1:
        query_data = query_data.reshape(-1, 1)

    # Select subset of query points
    np.random.seed(42)  # For reproducibility
    query_indices = np.random.choice(len(query_data), min(num_queries, len(query_data)), replace=False)
    query_points = query_data[query_indices]

    print(f"\nBenchmarking with {len(query_points)} queries...")
    print(f"Testing k values: {k_values}")
    print(f"Data dimensions - Train: {train_data.shape}, Query: {query_points.shape}")
    print(f"Using discriminator: {discriminator}")

    # Build and save initial tree
    print("\nBuilding and saving initial tree...")
    try:
        tree = KDTree(points=train_data, leafsize=leaf_size, discriminator=discriminator)
        tree_filename = "kdtree_benchmark.pkl"
        if not save_tree_to_disk(tree, tree_filename):
            print("Failed to save tree. Aborting benchmark.")
            return None
    except Exception as e:
        print(f"Error building initial tree: {e}")
        return None

    # Benchmark for each k value
    for k in k_values:
        print(f"\nTesting k={k}")
        gc.collect()  # Clear memory

        try:
            # Test rebuilding approach
            start_total = time.time()
            tree = KDTree(points=train_data, leafsize=leaf_size, discriminator=discriminator)
            start_query = time.time()

            # Process queries one by one since we're using k_nearest_neighbors
            for query in query_points:
                _ = tree.k_nearest_neighbors(query, k)

            end_query = time.time()

            rebuild_total_time = end_query - start_total
            rebuild_query_time = end_query - start_query

            # Test loading approach
            start_total = time.time()
            loaded_tree = load_tree_from_disk(tree_filename)
            if loaded_tree is None:
                print("Failed to load tree. Skipping load tests.")
                continue

            start_query = time.time()

            # Process queries one by one
            for query in query_points:
                _ = loaded_tree.k_nearest_neighbors(query, k)

            end_query = time.time()

            load_total_time = end_query - start_total
            load_query_time = end_query - start_query

            # Store results
            results['rebuild_times'].append(rebuild_total_time)
            results['load_times'].append(load_total_time)
            results['rebuild_query_times'].append(rebuild_query_time)
            results['load_query_times'].append(load_query_time)

            print(f"k={k} Results:")
            print(f"Rebuild + Query time: {rebuild_total_time:.4f} s")
            print(f"Load + Query time: {load_total_time:.4f} s")
            print(f"Rebuild Query time only: {rebuild_query_time:.4f} s")
            print(f"Load Query time only: {load_query_time:.4f} s")
            print(f"Average query time (rebuild): {rebuild_query_time/len(query_points):.6f} s")
            print(f"Average query time (load): {load_query_time/len(query_points):.6f} s")

        except Exception as e:
            print(f"Error during benchmark for k={k}: {e}")
            continue

    # Clean up
    try:
        if os.path.exists(tree_filename):
            os.remove(tree_filename)
    except Exception as e:
        print(f"Error cleaning up temporary file: {e}")

    return results

def plot_benchmark_results(results):
    """
    Plot benchmark results comparing rebuild vs load times.

    Args:
        results: Dictionary containing benchmark results
    """
    if not results:
        print("No results to plot")
        return

    k_values = results['k_values']

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

    # Plot total times
    ax1.plot(k_values, results['rebuild_times'], 'o-', label='Rebuild + Query')
    ax1.plot(k_values, results['load_times'], 's-', label='Load + Query')
    ax1.set_xlabel('k (number of neighbors)')
    ax1.set_ylabel('Total Time (seconds)')
    ax1.set_title('Total Operation Time vs k')
    ax1.legend()
    ax1.grid(True)

    # Plot query times only
    ax2.plot(k_values, results['rebuild_query_times'], 'o-', label='Query after Rebuild')
    ax2.plot(k_values, results['load_query_times'], 's-', label='Query after Load')
    ax2.set_xlabel('k (number of neighbors)')
    ax2.set_ylabel('Query Time (seconds)')
    ax2.set_title('Query Time vs k')
    ax2.legend()
    ax2.grid(True)

    plt.tight_layout()
    plt.show()
```
