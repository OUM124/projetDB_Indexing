```python
import numpy as np
class Rectangle:
    def __init__(self, *coords):
        self.coords = coords  # coordinates will be a tuple of size n (e.g., (x1, y1, z1, ..., xn), (x2, y2, z2, ..., xn))

    def area(self):
        # Calculate the volume (or area) of the bounding box in n dimensions
        volume = 1
        for i in range(len(self.coords)//2):
            volume *= self.coords[i + len(self.coords)//2] - self.coords[i]
        return volume

    def intersect(self, other):
        # Check if two rectangles (MBRs) intersect in n-dimensional space
        for i in range(len(self.coords)//2):
            if self.coords[i + len(self.coords)//2] < other.coords[i] or self.coords[i] > other.coords[i + len(other.coords)//2]:
                return False
        return True

    def merge(self, other):
        # Merge this rectangle with another in n-dimensional space
        merged_coords = []
        for i in range(len(self.coords)//2):
            merged_coords.append(min(self.coords[i], other.coords[i]))
            merged_coords.append(max(self.coords[i + len(self.coords)//2], other.coords[i + len(other.coords)//2]))
        return Rectangle(*merged_coords)

    def distance_to(self, point):
        distance = 0
        point = np.array(point)
        min_coords = np.array(self.coords[:len(self.coords)//2])
        max_coords = np.array(self.coords[len(self.coords)//2:])

        # Vectorized distance calculation
        below_min = point < min_coords
        above_max = point > max_coords

        dist_to_min = (min_coords - point) * below_min
        dist_to_max = (point - max_coords) * above_max

        total_dist = np.sum(dist_to_min**2 + dist_to_max**2)
        return np.sqrt(total_dist)
    def __repr__(self):
        return f"Rectangle({', '.join(map(str, self.coords))})"
```

```python
class LeafNode:
    def __init__(self, max_children, dimension, parent=None):
        self.is_leaf = True
        self.max_children = max_children
        self.entries = []  # List of data points
        self.mbr = None  # MBR of all the entries in this leaf node
        self.dimension = dimension  # Splitting dimension
        self.parent = parent  # Parent node (for underfill handling)

    def update_mbr(self):
        if self.entries:
            # Extract MBRs from the entries
            first_rectangle = self.entries[0][0]  # Get the rectangle from the first entry
            min_coords = list(first_rectangle.coords[:len(first_rectangle.coords)//2])
            max_coords = list(first_rectangle.coords[len(first_rectangle.coords)//2:])

            # Calculate the MBR by iterating over all rectangles
            for rectangle, _ in self.entries[1:]:
                for i in range(len(rectangle.coords)//2):
                    min_coords[i] = min(min_coords[i], rectangle.coords[i])
                    max_coords[i] = max(max_coords[i], rectangle.coords[i + len(rectangle.coords)//2])

            self.mbr = (min_coords, max_coords)


    def __str__(self):
        return f"LeafNode(mbr={self.mbr}, entries={self.entries})"

    # Add these methods to both InternalNode and LeafNode classes

    def quadratic_split(self):
        """
        Implements the quadratic split algorithm for R-tree nodes.
        Returns two nodes containing the split entries.
        """
        # Helper function to calculate area enlargement
        def calc_enlargement(mbr1, mbr2):
            merged_min = [min(mbr1[0][i], mbr2[0][i]) for i in range(len(mbr1[0]))]
            merged_max = [max(mbr1[1][i], mbr2[1][i]) for i in range(len(mbr1[0]))]

            # Calculate original areas
            area1 = Rectangle(*mbr1[0], *mbr1[1]).area()
            area2 = Rectangle(*mbr2[0], *mbr2[1]).area()

            # Calculate merged area
            merged_area = Rectangle(*merged_min, *merged_max).area()
            return merged_area - (area1 + area2)

        # Find the two entries that would waste the most area if put together
        max_waste = float('-inf')
        seed1, seed2 = 0, 0

        for i in range(len(self.entries)):
            mbr1 = self.entries[i][1] if not self.is_leaf else (self.entries[i][0].coords[:len(self.entries[i][0].coords)//2],
                                                              self.entries[i][0].coords[len(self.entries[i][0].coords)//2:])
            for j in range(i + 1, len(self.entries)):
                mbr2 = self.entries[j][1] if not self.is_leaf else (self.entries[j][0].coords[:len(self.entries[j][0].coords)//2],
                                                                  self.entries[j][0].coords[len(self.entries[j][0].coords)//2:])
                waste = calc_enlargement(mbr1, mbr2)
                if waste > max_waste:
                    max_waste = waste
                    seed1, seed2 = i, j

        # Create two new nodes and distribute entries
        new_node1 = type(self)(self.max_children, self.dimension, self.parent)
        new_node2 = type(self)(self.max_children, self.dimension, self.parent)

        # Initialize with seed entries
        new_node1.entries = [self.entries[seed1]]
        new_node2.entries = [self.entries[seed2]]

        # Remove seeds from entries to distribute
        remaining_entries = [entry for i, entry in enumerate(self.entries) if i not in {seed1, seed2}]

        # Distribute remaining entries
        while remaining_entries and (len(new_node1.entries) < self.max_children - 1 or
                                  len(new_node2.entries) < self.max_children - 1):
            if len(remaining_entries) == 1:
                # Add last entry to the group with fewer entries
                if len(new_node1.entries) < len(new_node2.entries):
                    new_node1.entries.append(remaining_entries[0])
                else:
                    new_node2.entries.append(remaining_entries[0])
                break

            # Find entry with greatest preference for one group
            max_diff = float('-inf')
            chosen_entry = 0
            chosen_group = None

            for i, entry in enumerate(remaining_entries):
                mbr = entry[1] if not self.is_leaf else (entry[0].coords[:len(entry[0].coords)//2],
                                                        entry[0].coords[len(entry[0].coords)//2:])

                # Calculate enlargement needed for both groups
                d1 = calc_enlargement(new_node1.get_mbr(), mbr)
                d2 = calc_enlargement(new_node2.get_mbr(), mbr)

                diff = abs(d1 - d2)
                if diff > max_diff:
                    max_diff = diff
                    chosen_entry = i
                    chosen_group = 1 if d1 < d2 else 2

            # Add entry to chosen group
            if chosen_group == 1:
                new_node1.entries.append(remaining_entries[chosen_entry])
            else:
                new_node2.entries.append(remaining_entries[chosen_entry])

            remaining_entries.pop(chosen_entry)

        # Update MBRs
        new_node1.update_mbr()
        new_node2.update_mbr()

        return new_node1, new_node2

    def linear_split(self):
        """
        Implements the linear split algorithm for R-tree nodes.
        Returns two nodes containing the split entries.
        """
        # Find the dimension with the greatest normalized separation
        best_dim = 0
        best_sep = float('-inf')

        for dim in range(self.dimension):
            # Sort entries by their lower and upper bounds in this dimension
            if self.is_leaf:
                lower_sorted = sorted(self.entries, key=lambda x: x[0].coords[dim])
                upper_sorted = sorted(self.entries, key=lambda x: x[0].coords[dim + len(x[0].coords)//2])
            else:
                lower_sorted = sorted(self.entries, key=lambda x: x[1][0][dim])
                upper_sorted = sorted(self.entries, key=lambda x: x[1][1][dim])

            # Calculate the separation
            width = (upper_sorted[-1][1][1][dim] if not self.is_leaf else upper_sorted[-1][0].coords[dim + len(upper_sorted[-1][0].coords)//2]) - \
                    (lower_sorted[0][1][0][dim] if not self.is_leaf else lower_sorted[0][0].coords[dim])

            if width == 0:
                continue

            # Calculate normalized separation
            sep = abs(upper_sorted[-1][1][1][dim] - lower_sorted[-1][1][0][dim]) / width if not self.is_leaf else \
                  abs(upper_sorted[-1][0].coords[dim + len(upper_sorted[-1][0].coords)//2] - lower_sorted[-1][0].coords[dim]) / width

            if sep > best_sep:
                best_sep = sep
                best_dim = dim

        # Sort entries by their center in the chosen dimension
        if self.is_leaf:
            sorted_entries = sorted(self.entries,
                                  key=lambda x: (x[0].coords[best_dim] + x[0].coords[best_dim + len(x[0].coords)//2]) / 2)
        else:
            sorted_entries = sorted(self.entries,
                                  key=lambda x: (x[1][0][best_dim] + x[1][1][best_dim]) / 2)

        # Split at the median
        split_point = len(sorted_entries) // 2

        # Create and populate the new nodes
        new_node1 = type(self)(self.max_children, self.dimension, self.parent)
        new_node2 = type(self)(self.max_children, self.dimension, self.parent)

        new_node1.entries = sorted_entries[:split_point]
        new_node2.entries = sorted_entries[split_point:]

        # Update MBRs
        new_node1.update_mbr()
        new_node2.update_mbr()

        return new_node1, new_node2

    def get_mbr(self):
        """Helper method to get current MBR of a node's entries"""
        if not self.entries:
            return None

        if self.is_leaf:
            min_coords = list(self.entries[0][0].coords[:len(self.entries[0][0].coords)//2])
            max_coords = list(self.entries[0][0].coords[len(self.entries[0][0].coords)//2:])

            for entry in self.entries[1:]:
                for i in range(len(min_coords)):
                    min_coords[i] = min(min_coords[i], entry[0].coords[i])
                    max_coords[i] = min(max_coords[i], entry[0].coords[i + len(entry[0].coords)//2])
        else:
            min_coords = list(self.entries[0][1][0])
            max_coords = list(self.entries[0][1][1])

            for entry in self.entries[1:]:
                for i in range(len(min_coords)):
                    min_coords[i] = min(min_coords[i], entry[1][0][i])
                    max_coords[i] = max(max_coords[i], entry[1][1][i])

        return (min_coords, max_coords)
```

```python
import heapq
import math

class RTree:
    def __init__(self, max_children=4, min_children=2, dimension=2):
        self.root = LeafNode(max_children, dimension)
        self.max_children = max_children
        self.min_children = min_children
        self.dimension = dimension  # n-dimensional space
       # print(f"RTree initialized with max_children={max_children}, min_children={min_children}, dimension={dimension}")

    def insert(self, point):
        # Create a Rectangle for the point (point is a tuple of coordinates)
        rectangle = Rectangle(*point, *point)  # point is a tuple of coordinates
       # print(f"Inserting point {point} with MBR {rectangle}")

        leaf = self._choose_leaf(self.root, rectangle)
        #print(f"Chosen leaf for insertion: {leaf}")

        leaf.entries.append((rectangle, point))  # Store point with its MBR
        leaf.update_mbr()
       # print(f"Updated MBR of leaf after insertion: {leaf.mbr}")

        if len(leaf.entries) > self.max_children:
            self._split_node(leaf)

    def _choose_leaf(self, node, rectangle):
        if node.is_leaf:
            return node

        # For internal nodes, choose the child with the least enlargement of its MBR
        best_child = None
        best_enlargement = float('inf')

        for child, child_mbr in node.entries:
            enlargement = self._calculate_enlargement(child_mbr, rectangle)
            #print(f"Calculating enlargement for MBR {child_mbr} with new rectangle {rectangle}: {enlargement}")

            if enlargement < best_enlargement:
                best_enlargement = enlargement
                best_child = child

        #print(f"Best child chosen: {best_child}")
        return self._choose_leaf(best_child, rectangle)

    def _calculate_enlargement(self, mbr, rectangle):
        # `mbr` is a tuple of (min_coords, max_coords)
        mbr_min, mbr_max = mbr
        rect_min, rect_max = rectangle.coords[:len(rectangle.coords)//2], rectangle.coords[len(rectangle.coords)//2:]

        # Calculate enlargement in each dimension
        enlargement = 1
        for i in range(len(mbr_min)):
            enlarged_min = min(mbr_min[i], rect_min[i])
            enlarged_max = max(mbr_max[i], rect_max[i])
            enlargement *= (enlarged_max - enlarged_min)

        # Subtract the original MBR volume
        original_volume = 1
        for i in range(len(mbr_min)):
            original_volume *= (mbr_max[i] - mbr_min[i])

        return enlargement - original_volume


    def _split_node(self, node):
        #print(f"Splitting node with {len(node.entries)} entries")
        half = len(node.entries) // 2
        new_node = InternalNode(self.max_children, self.dimension) if not node.is_leaf else LeafNode(self.max_children, self.dimension)
        new_node.entries = node.entries[half:]
        node.entries = node.entries[:half]

        # Update the MBRs after split
        node.update_mbr()
        new_node.update_mbr()

       # print(f"Node split completed. Updated MBR of original node: {node.mbr}")
       # print(f"New node created with MBR: {new_node.mbr}")

        if node == self.root:
            #print("Splitting root node, creating a new root")
            new_root = InternalNode(self.max_children, self.dimension)
            new_root.entries = [(node, node.mbr), (new_node, new_node.mbr)]
            node.parent = new_root  # Set parent for the original node
            new_node.parent = new_root  # Set parent for the new node
            new_root.update_mbr()
            self.root = new_root
            #print(f"New root created with MBR: {self.root.mbr}")
        else:
            # Add the new node to the parent of the current node
            parent = node.parent
            if parent is None:
                raise ValueError("Parent should not be None during a non-root node split")
            new_node.parent = parent  # Set the parent for the new node
            parent.entries.append((new_node, new_node.mbr))
            parent.update_mbr()
            #print(f"Adding new node to parent: {parent}")

        # Enforce minimum entries for the node
        self._enforce_minimum_entries(node)

    def _enforce_minimum_entries(self, node):
        if len(node.entries) < self.min_children and node != self.root:
            #print(f"Node {node} has fewer than {self.min_children} entries, enforcing minimum entries")

            parent = node.parent
            for i, (child, child_mbr) in enumerate(parent.entries):
                if child == node:
                    if i > 0:  # Merge with previous sibling
                        sibling = parent.entries[i - 1][0]
                        #print(f"Merging with previous sibling: {sibling}")
                        sibling.entries.extend(node.entries)
                        sibling.update_mbr()
                        parent.entries.remove((node, node.mbr))
                    elif i < len(parent.entries) - 1:  # Merge with next sibling
                        sibling = parent.entries[i + 1][0]
                        #print(f"Merging with next sibling: {sibling}")
                        sibling.entries.extend(node.entries)
                        sibling.update_mbr()
                        parent.entries.remove((node, node.mbr))

                    parent.update_mbr()
                    #print(f"Parent MBR updated after merging: {parent.mbr}")
                    break

    def search(self, query_rectangle):
        return self._search_recursive(self.root, query_rectangle)

    def _search_recursive(self, node, query_rectangle):
        results = []

        if node.is_leaf:
            # Step 2: Search the leaf node
            for rect, point in node.entries:
                if rect.intersect(query_rectangle):  # `rect` is already a Rectangle object
                    results.append(point)
        else:
            # Step 1: Search subtrees for internal nodes
            for child, child_mbr in node.entries:
                child_rectangle = Rectangle(*child_mbr[0], *child_mbr[1])  # Convert tuple to Rectangle
                if child_rectangle.intersect(query_rectangle):
                    results.extend(self._search_recursive(child, query_rectangle))

        return results

    def nearest_neighbor(self, query_point):
        return self._nearest_neighbor_recursive(self.root, query_point, best=None)

    def _nearest_neighbor_recursive(self, node, query_point, best=None):
        if node.is_leaf:
            for rect, point in node.entries:
                dist = rect.distance_to(query_point)  # `rect` is already a Rectangle object
                if best is None or dist < best[0]:
                    best = (dist, point)
        else:
            # Search internal nodes (recursive)
            priority_queue = []
            for child, child_mbr in node.entries:
                child_rectangle = Rectangle(*child_mbr[0], *child_mbr[1])  # Convert tuple to Rectangle
                dist = child_rectangle.distance_to(query_point)  # Use Rectangle's distance_to method
                if best is None or dist < best[0]:
                    heapq.heappush(priority_queue, (dist, child))

            while priority_queue:
                _, child = heapq.heappop(priority_queue)
                best = self._nearest_neighbor_recursive(child, query_point, best)
                #print("best ", best)
        return best if best else None



    def bulk_load(self, points):
        """
        Bulk load the R-tree using the given points.
        Points are tuples representing coordinates in the space.
        """
        # Sort points along dimensions (for STR-like bulk loading)
        sorted_points = sorted(points, key=lambda p: p[0])  # Sort by the first dimension

        # Group points into tiles and create leaf nodes
        tile_size = self.max_children

        tiles = [sorted_points[i:i + tile_size] for i in range(0, len(sorted_points), tile_size)]

        leaf_nodes = []
        for tile in tiles:
            leaf_node = LeafNode(self.max_children, self.dimension)
            for point in tile:
                rect = Rectangle(*point, *point)
                leaf_node.entries.append((rect, point))
            leaf_node.update_mbr()
            leaf_nodes.append(leaf_node)

        # Build the tree from leaf nodes
        while len(leaf_nodes) > 1:
            parents = []
            for i in range(0, len(leaf_nodes), self.max_children):
                parent = InternalNode(self.max_children, self.dimension)
                children = leaf_nodes[i:i + self.max_children]
                for child in children:
                    parent.entries.append((child, child.mbr))
                    child.parent = parent
                parent.update_mbr()
                parents.append(parent)
            leaf_nodes = parents

        # Set the root of the tree

        self.root = leaf_nodes[0]


    def str_bulk_load(self, points):
        """
        Improved Sort-Tile-Recursive (STR) bulk loading algorithm.
        """
        print(f"Starting STR bulk loading with {len(points)} points")

        if len(points) == 0:
            raise ValueError("Cannot bulk load empty points list")

        def partition_points(points_subset, depth=0):
            """
            Partitions points into groups that will form nodes.
            Uses the STR approach of sorting and splitting by different dimensions.
            """
            if len(points_subset) <= self.max_children:
                return [points_subset]

            # Calculate number of splits needed
            n = len(points_subset)
            num_leaf_nodes = math.ceil(n / self.max_children)
            num_splits_per_dim = int(math.ceil(num_leaf_nodes ** (1.0 / 2)))  # Square root for 2D-like splits

            # Sort by current dimension
            dim = depth % self.dimension
            points_subset.sort(key=lambda x: x[dim])

            # Split into vertical slices
            slice_size = math.ceil(len(points_subset) / num_splits_per_dim)
            slices = []

            for i in range(0, len(points_subset), slice_size):
                slice_points = points_subset[i:i + slice_size]
                if len(slice_points) > self.max_children:
                    # Sort by next dimension and split again
                    next_dim = (depth + 1) % self.dimension
                    slice_points.sort(key=lambda x: x[next_dim])
                    inner_slice_size = math.ceil(len(slice_points) / num_splits_per_dim)
                    for j in range(0, len(slice_points), inner_slice_size):
                        partition = slice_points[j:j + inner_slice_size]
                        if partition:  # Only add non-empty partitions
                            slices.append(partition)
                else:
                    if slice_points:  # Only add non-empty slices
                        slices.append(slice_points)

            print(f"Created {len(slices)} partitions at depth {depth}, avg size: {sum(len(s) for s in slices)/len(slices):.1f}")
            return slices

        def build_leaves(points_subset):
            """Creates leaf nodes from point partitions"""
            leaf_nodes = []
            partitions = partition_points(points_subset)

            for partition in partitions:
                leaf = LeafNode(self.max_children, self.dimension)
                for point in partition:
                    rect = Rectangle(*point, *point)
                    leaf.entries.append((rect, point))
                leaf.update_mbr()
                leaf_nodes.append(leaf)

            print(f"Created {len(leaf_nodes)} leaf nodes")
            return leaf_nodes

        def build_tree(nodes, level=0):
            """Builds the tree bottom-up from the leaf nodes"""
            if len(nodes) <= 1:
                return nodes[0]

            # Group nodes into internal nodes
            internal_nodes = []
            for i in range(0, len(nodes), self.max_children):
                group = nodes[i:i + self.max_children]
                internal = InternalNode(self.max_children, self.dimension)
                for child in group:
                    child.parent = internal
                    internal.entries.append((child, child.mbr))
                internal.update_mbr()
                internal_nodes.append(internal)

            print(f"Created {len(internal_nodes)} internal nodes at level {level}")
            return build_tree(internal_nodes, level + 1)

        # Build the tree
        leaf_nodes = build_leaves(list(points))  # Convert to list if not already
        self.root = build_tree(leaf_nodes)
        print("STR bulk loading completed")
    def _knn_recursive(self, node, query_point, k, best):
        """Optimized KNN recursive search"""
        if node.is_leaf:
            for rect, point in node.entries:
                dist = rect.distance_to(query_point)

                if len(best) < k:
                    # Convert point to tuple only if it's a numpy array
                    point_tuple = tuple(point) if hasattr(point, '__array__') else point
                    heapq.heappush(best, (-dist, point_tuple))
                else:
                    max_dist = -best[0][0]
                    if dist < max_dist:
                        point_tuple = tuple(point) if hasattr(point, '__array__') else point
                        heapq.heappushpop(best, (-dist, point_tuple))
        else:
            # Process internal nodes
            entries = [
                (Rectangle(*child_mbr[0], *child_mbr[1]).distance_to(query_point), child)
                for child, child_mbr in node.entries
            ]
            entries.sort(key=lambda x: x[0])  # Sort by distance

            for min_dist, child in entries:
                if best and len(best) >= k:
                    if min_dist > -best[0][0]:  # Early pruning
                        break
                self._knn_recursive(child, query_point, k, best)

    def knn(self, query_point, k):
        """Optimized k-nearest neighbors search"""
        # Convert query_point to tuple only if it's a numpy array
        if hasattr(query_point, '__array__'):
            query_point = tuple(query_point)

        best = []
        self._knn_recursive(self.root, query_point, k, best)
        return [point for _, point in sorted(best, key=lambda x: abs(x[0]))]
```

```python
import numpy as np
from collections import defaultdict

def create_vector_to_index_mapping(train_data, decimals=10):
    """
    Create a mapping from vector tuple to index for fast lookup.

    Args:
        train_data: numpy array of shape (1000000, 960)
        decimals: number of decimal places to round to
    """
    # Round the data to handle floating point precision
    rounded_data = np.round(train_data, decimals=decimals)

    # Create mapping of vector to index
    vector_to_idx = {}
    for idx, vector in enumerate(rounded_data):
        vector_to_idx[tuple(vector)] = idx

    return vector_to_idx

def batch_find_indices(vectors, vector_to_idx_map, decimals=10):
    """
    Find indices for multiple vectors at once using the precomputed mapping.

    Args:
        vectors: numpy array of vectors
        vector_to_idx_map: dict mapping vector tuples to indices
        decimals: number of decimal places to round to
    """
    rounded_vectors = np.round(vectors, decimals=decimals)
    indices = []
    for v in rounded_vectors:
        indices.append(vector_to_idx_map.get(tuple(v), -1))
    return np.array(indices)

def evaluate_knn_accuracy_optimized(knn_results, ground_truth_indices, train_data):
    """
    Optimized version of KNN accuracy evaluation.

    Args:
        knn_results: numpy array of shape (num_queries, num_neighbors) containing neighbor vectors
        ground_truth_indices: numpy array of shape (num_queries, num_neighbors) containing true indices
        train_data: numpy array of shape (num_train, vector_dim) containing training data
    """
    # Create vector to index mapping once
    vector_to_idx_map = create_vector_to_index_mapping(train_data)

    num_queries = len(knn_results)
    num_neighbors = len(knn_results[0])

    # Process each query point
    predicted_indices = []
    for query_neighbors in knn_results:
        # Find indices for all neighbors of this query point
        query_indices = batch_find_indices(query_neighbors, vector_to_idx_map)
        predicted_indices.append(query_indices)

    predicted_indices = np.array(predicted_indices)

    # Compute accuracy using vectorized operations
    correct_predictions = np.sum([
        np.isin(ground_truth_indices[i], predicted_indices[i]).sum()
        for i in range(num_queries)
    ])

    accuracy = correct_predictions / (num_queries * num_neighbors)
    return accuracy

```
